## CNN Feature Extraction

```python
import numpy as np
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
```

```python
import os
```

```python
from tensorflow.keras.models import Model
```

```python
base_model = ResNet50(weights='imagenet')
```

```python
model_conv1 = Model(inputs=base_model.input, outputs=base_model.get_layer('pool1_pool').output)
```

```python
stripe_resnet_conv1_embeddings = {}
```

```python
folder_dir = "Convolutional DNN/"
i = 0
image_count = 1000
for folder in os.listdir(folder_dir):
    if ".DS_" not in folder:
        for filename in os.listdir(folder_dir + folder):
            if filename.endswith(".png"):
                print('Processing image {}/{}'.format( i+1, image_count ))
                img_path = folder_dir + folder + "/" + filename
                img = image.load_img(img_path, target_size=(224, 224))
                x = image.img_to_array(img)
                x = np.expand_dims(x, axis=0)
                x = preprocess_input(x)
                vec = model_conv1.predict(x)[0][55][55]
                stripe_resnet_conv1_embeddings[filename] = vec
                i += 1
```

```python
model_conv3 = Model(inputs=base_model.input, outputs=base_model.get_layer('conv3_block4_out').output)
```

```python
stripe_resnet_conv3_embeddings = {}
```

```python
folder_dir = "Convolutional DNN/"
i = 0
image_count = 1000
for folder in os.listdir(folder_dir):
    if ".DS_" not in folder:
        for filename in os.listdir(folder_dir + folder):
            if filename.endswith(".png"):
                print('Processing image {}/{}'.format( i+1, image_count ))
                img_path = folder_dir + folder + "/" + filename
                img = image.load_img(img_path, target_size=(224, 224))
                x = image.img_to_array(img)
                x = np.expand_dims(x, axis=0)
                x = preprocess_input(x)
                vec = model_conv3.predict(x)[0][27][27]
                stripe_resnet_conv3_embeddings[filename] = vec
                i += 1
```

```python
model_conv5 = Model(inputs=base_model.input, outputs=base_model.get_layer('conv5_block3_out').output)
```

```python
stripe_resnet_conv5_embeddings = {}
```

```python
folder_dir = "Convolutional DNN/"
i = 0
image_count = 1000
for folder in os.listdir(folder_dir):
    if ".DS_" not in folder:
        for filename in os.listdir(folder_dir + folder):
            if filename.endswith(".png"):
                print('Processing image {}/{}'.format( i+1, image_count ))
                img_path = folder_dir + folder + "/" + filename
                img = image.load_img(img_path, target_size=(224, 224))
                x = image.img_to_array(img)
                x = np.expand_dims(x, axis=0)
                x = preprocess_input(x)
                vec = model_conv5.predict(x)[0][6][6]
                stripe_resnet_conv5_embeddings[filename] = vec
                i += 1
```

```python
import pickle
```

```python
with open('stripe_resnet_conv1_embeddings.pickle', 'wb') as handle:
    pickle.dump(stripe_resnet_conv1_embeddings, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('stripe_resnet_conv3_embeddings.pickle', 'wb') as handle:
    pickle.dump(stripe_resnet_conv3_embeddings, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('stripe_resnet_conv5_embeddings.pickle', 'wb') as handle:
    pickle.dump(stripe_resnet_conv5_embeddings, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
